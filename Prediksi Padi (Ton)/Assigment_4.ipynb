**Import Library**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, root_mean_squared_error
```

**Load Data**

```python
data_set = '/content/produksiPadi.xlsx'
df = pd.read_excel(data_set, usecols = ['Tahun','Produksi Padi(Ton)'])
```

```python
df.head()
```

```python
# Informasi umum tentang data
df.info()
```

```python
# Statistik deskriptif
df.describe()
```

```python
# Cek nilai kosong
df.isnull().sum()
```

```python
# Membuat kolom baru untuk menyimapan nilai-nilai produksi dari padi
df['Produksi_Tahun_Sebelumnya'] = df['Produksi Padi(Ton)'].shift(1)
```

```python
df.head()
```

```python
# Menangani missing value pada kolom "Produksi_Tahun_Sebelumnya"
df['Produksi_Tahun_Sebelumnya'].fillna(df['Produksi_Tahun_Sebelumnya'].mode()[0], inplace = True)
```

```python
df.head()
```

```python
# Membagi data menjadi 2 (x = fitur, y = target)
x = df['Produksi_Tahun_Sebelumnya'].values.reshape(-1,1)
y = df['Produksi Padi(Ton)'].values.reshape(-1,1)
```

```python
# Membagi dataset menjadi 2 (training dan testing)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
ln_reg = LinearRegression()
```

```python
# Melatih data dengan model Linear Regression
ln_reg.fit(x_train, y_train)
```

```python
# Prediksi dengan model Linear Regression
y_pred = ln_reg.predict(x_test)
```

```python
# Menghitung akurasi
accurate = ln_reg.score(x_test, y_test) *100
accurate
```

```python
# Menghitung metrik evaluasi dalam Regression
MSE = mean_squared_error(y_test, y_pred)
RMS = root_mean_squared_error(y_test, y_pred)
MAPE = mean_absolute_percentage_error(y_test, y_pred)
```
